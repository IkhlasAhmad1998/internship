```python
import os
from dotenv import load_dotenv, find_dotenv
from pypdf import PdfReader
import google.generativeai as genai
from pprint import pprint
```

```python
# reading the pdf
reader = PdfReader("data/Labor Law.pdf")

data = []
for page in reader.pages:
    text = page.extract_text()
    data.append(text)
```

```python
load_dotenv(find_dotenv(".env"))

GOOGLE_GENAI_API_KEY = os.getenv("GEMINI_API_KEY")

genai.configure(api_key=GOOGLE_GENAI_API_KEY)
```

```python
model = genai.GenerativeModel("gemini-1.5-flash-latest")
```

```python
context = "\n".join(data)
```

```python
prompt = (f"""أرغب في الحصول على ملخص مفصل وشامل لمحتويات هذه الوثيقة، التي تتناول نظام العمل في المملكة العربية السعودية. يجب أن يغطي الملخص جميع جوانب الوثيقة، بما في ذلك الأحكام القانونية، والتعريفات، والمواد التنظيمية، والحقوق والواجبات، والعقوبات أو الإجراءات القانونية، إن وجدت.

يجب أن يكون الملخص واضحًا وسهل الفهم، ويمنح القارئ فهمًا دقيقًا وشاملًا لمحتوى الوثيقة، حتى لو لم يكن متخصصًا قانونيًا.

هام جدًا: يجب حفظ جميع المصطلحات القانونية كما وردت دون ترجمة أو تبسيط، حتى لا يُغير ذلك من المعنى القانوني. يُرجى عدم اختصار أو حذف أي جزء ضروري لفهم القانون.

يرجى تقديم الملخص باللغة العربية الفصحى.

نص الوثيقة:

{context }""")

response = model.generate_content(prompt)
```

```python
pprint(response.text)
```
